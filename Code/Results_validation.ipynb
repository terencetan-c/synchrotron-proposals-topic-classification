```python
def compare_predictions(row):
    a = row['topic_predictions_pdf_metadata_only']
    b = row['topic_predictions_combined']

    c = {item['topic_id'] for item in a}  # Set from a
    d = {item['topic_id'] for item in b}  # Set from b
    res = c.intersection(d)  # Common IDs
    if (len(res) != len(a)) or (len(res) != len(b)):
        return len(res)
```

```python
prop_for_val[prop_for_val.apply(compare_predictions, axis=1).notna()][['topic_predictions_pdf_metadata_only','topic_predictions_combined']]
```
