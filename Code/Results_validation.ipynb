```python
prop_esrf_pdf_only=prop_esrf_pdf_only.rename(columns={'title_y':'title','abstract_inverted_index_y':'abstract_inverted_index','referenced_works_pdf':'referenced_works','journal_display_name_y':'journal_display_name','Experiment session DOI_y':'experiment session DOI','referenced_works DOI_y':'referenced_works DOI'})
prop_esrf_pdf_only['abstract_inverted_index'].fillna('',inplace=True)
prop_esrf_pdf_only['title'].fillna('',inplace=True)
```

```python
prop_esrf_pdf_only
```

```python
batch=prop_esrf_pdf_only[['title','abstract_inverted_index','referenced_works','journal_display_name']].to_dict(orient='records')
for dict in batch:
    dict['inverted']=False
```

```python
# Proceed with topic classification using the full model
results=[]
for single in batch:
    result=predictor.full_model_prediction([single])
    results.append(json.loads(result)[0])
```

```python
# Add the list of topic predictions as a dict value to the batch

n = len(batch)
for i in range(n):
    batch[i]['topic predictions']=results[i]
    batch[i]['Proposal number']=prop_esrf_pdf_only['Proposal number'].to_list()[i]
    batch[i]['Experiment session DOI']=prop_esrf_pdf_only['experiment session DOI'].to_list()[i]
    batch[i]['PDF document name']=prop_esrf_pdf_only['PDF document name'].to_list()[i]
    batch[i]['referenced_works DOI']=prop_esrf_pdf_only['referenced_works DOI'].to_list()[i]
```

```python
# Export predictions as a JSON file

with open('/Users/fdp54928/Library/CloudStorage/OneDrive-Nexus365/GitHub Repositories/synchrotron-proposals-topic-classification/Datasets/ESRF/Predictions/Proposals_ESRF_Predictions_with_PDF_metadata_only', 'w') as f:
    json.dump(batch, f, indent=2)
```
